```python
#1. 안전한 정수 입력 함수
# 사용자에게 숫자를 입력받는 함수 `input_int(prompt)`를 작성하세요.
#1. 정수가 아니면 `"정수를 입력하세요."`를 출력하세요.
#2. 올바른 정수를 입력할 때까지 반복하세요.
#3. 정상 입력 시 `int` 값을 반환하세요.

def input_int(prompt):
    while True:
        try: 
            user_input = input(prompt)
            return int(user_input)
        except ValueError:
            print("정수를 입력하세요.")

print(input_int("숫자를 입력해주세요"))
```

```python
#2. 두 정수를 입력받아 `a / b` 결과를 출력하세요.
#1. 0으로 나누면 `"0으로 나눌 수 없습니다."`를 출력하세요.
#2. 정수가 아닌 입력이면 `"정수만 입력하세요."`를 출력하세요.
#3. 어떤 경우든 프로그램이 종료되지 않고 다시 입력받도록 작성하세요.

def input_int():
    while True:
        try: 
            a = int(input("첫번째 숫자를 입력해주세요."))
            b = int(input("2번째 숫자를 입력해주세요."))
            return a/b
        except ValueError:
            print("정수만 입력하세요.")
        except ZeroDivisionError:
            print("0으로 나눌 수 없습니다.")

print(input_int())
```

```python
### 문제 4. 사용자 정의 예외로 규칙 강제
##점수 입력 규칙: **0~100 사이 정수만 허용**
#1. 규칙을 위반하면 `InvalidScoreError` 예외를 발생시키세요.
#2. main 로직에서 해당 예외를 잡아 `"점수는 0~100 사이여야 합니다."`를 출력하세요.
#3. 올바른 점수를 입력할 때까지 반복하세요.

class InvalidScoreError(Exception):
    def __init__(self):
        super().__init__("점수는 0~100점 사이여야 합니다.")

# 2. 예외 발생시키기 (raise 사용)
def check_positive(number):
    if number < 0 or number > 100:
        raise InvalidScoreError() # raise는 파이썬에게 "자, 여기서 에러가 났으니까 당장 멈추고 처리할 곳(except)을 찾아!"라고 명령하는 도구입니다.
    return number

def input_int():
    while True:
        try:
            num = int(input("시험점수를 입력하세요: "))
            check_positive(num)
            return num
        except InvalidScoreError as e:
            print(f"에러 발생: {e}")
        except ValueError:
            print("정수만 입력하세요.")

print(f"시험점수 : {input_int()}")
```

```python
#5번 문제

def input_int():
    while True:
        try: 
            a = int(input("첫번째 숫자를 입력해주세요."))
            b = int(input("2번째 숫자를 입력해주세요."))
            return a/b
        except Exception:
        # 예외가 발생했을 때 실행
            print("오류가 발생했습니다.")
        finally:
        # 성공하든 실패하든 무조건 실행
            print("프로그램을 종료합니다.")
            break

print(input_int())
```
