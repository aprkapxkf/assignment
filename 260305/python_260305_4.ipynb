```python
# 3번 문제
class MenuItem:
    def __init__(self, name, unit_price):
        self.name = name
        self.unit_price = unit_price

    def get_info(self):
        return f"{self.name} - {self.unit_price}원"

class Coffee(MenuItem):
    def __init__(self, name, unit_price, shot=1):
        super().__init__(name, unit_price)
        self.shot = shot

    def get_info(self):
        return f"{self.name}(샷 {self.shot}) - {self.unit_price}원"

class Dessert(MenuItem):
    def __init__(self, name, unit_price, size="M"):
        super().__init__(name, unit_price)
        self.size = size

    def get_info(self):
        return f"{self.name}({self.size}) - {self.unit_price}원"

items = [
    Coffee("아메리카노", 3000, 2),
    Dessert("치즈케이크", 5500, "L"),
    Coffee("에스프레소", 2500)  # 기본 샷 1 적용
]

print("--- 카페 메뉴 정보 ---")
for item in items:

    print(item.get_info())
```

```python
class OrderItem:
    def __init__(self, name, unit_price, qty):
        self.name = name
        self.unit_price = unit_price
        self.qty = qty

    def total_price(self):
        return self.unit_price * self.qty

    def __str__(self):
        return f"{self.name} x{self.qty} = {self.total_price()}원"


class Cart:
    def __init__(self):
        self.items = []

    def add(self, item):
        self.items.append(item)

    def total(self):
        return sum(item.total_price() for item in self.items)

    def receipt(self):
        for item in self.items:
            print(item)
        print(f"총합: {self.total()}원")


cart = Cart()
cart.add(OrderItem("아메리카노", 3000, 2))
cart.add(OrderItem("치즈케이크", 5500, 1))

cart.receipt()
```

```python
from abc import ABC, abstractmethod

# [1] 주문 항목 클래스: 개별 상품의 데이터와 계산 로직을 담당
class OrderItem:
    def __init__(self, name, unit_price, qty):
        self.name = name           # 상품명
        self.unit_price = unit_price # 단가
        self.qty = qty             # 수량

    def total_price(self):
        """해당 항목의 총 가격(단가 * 수량)을 계산하여 반환"""
        return self.unit_price * self.qty

    def __str__(self):
        """객체를 print()할 때 보여줄 문자열 형식 정의"""
        return f"{self.name} x{self.qty} = {self.total_price()}원"


# [2] 장바구니 클래스: 상품들을 담고 결제 과정을 조율함 (Context 역할)
class Cart:
    def __init__(self):
        self.items = [] # OrderItem 객체들을 담을 리스트

    def add(self, item):
        """장바구니에 주문 항목 추가"""
        self.items.append(item)

    def total(self):
        """장바구니에 담긴 모든 항목의 합계 금액 계산"""
        # 리스트 내포와 sum 함수를 사용하여 간결하게 합산
        return sum(item.total_price() for item in self.items)

    def checkout(self, payment):
        """
        결제 진행 (전략 패턴의 핵심)
        외부에서 주입된 payment 객체가 무엇이든 상관없이 .pay()를 호출함
        """
        amount = self.total()
        payment.pay(amount)


# [3] 결제 추상 클래스: 모든 결제 수단의 부모 (Strategy 인터페이스 역할)
class Payment(ABC):
    @abstractmethod
    def pay(self, amount):
        """자식 클래스에서 반드시 구현해야 하는 추상 메서드"""
        pass


# [4] 구체적인 결제 전략 1: 카드 결제
class CardPayment(Payment):
    def pay(self, amount):
        print(f"카드로 {amount}원 결제 완료")


# [5] 구체적인 결제 전략 2: 현금 결제
class CashPayment(Payment):
    def pay(self, amount):
        print(f"현금으로 {amount}원 결제 완료")


# --- 실제 실행 부분 ---

# 1. 장바구니 생성 및 상품 추가
cart = Cart()
cart.add(OrderItem("아메리카노", 3000, 2))
cart.add(OrderItem("치즈케이크", 5500, 1))

# 2. 카드 결제 전략을 주입하여 결제 실행
# Cart는 내부 로직 변경 없이 CardPayment의 기능을 사용함
cart.checkout(CardPayment())

# 3. 현금 결제 전략을 주입하여 결제 실행
cart.checkout(CashPayment())
```
