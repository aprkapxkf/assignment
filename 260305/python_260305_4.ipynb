```python
# 1번 문제

class Orderitem():
    def __init__(self, name, unit_price, qty):
        self.name = name
        self.unit_price = unit_price
        self.qty = qty
    def total_price(self):
        t = self.unit_price * self.qty
        return t
    def __str__(self):
        return f"{self.name} * {self.qty} = {self.total_price()}원"
    
item1 = Orderitem("아메리카노", 3000, 2)
item2 = Orderitem("치즈케이크", 5000, 1)

print(item1)
print(item2)
total_sum = item1.total_price() + item2.total_price()
print(f"총합: {total_sum}원")
```

```python
# 2번 문제
class Orderitem():
    def __init__(self, name, __unit_price, __qty):
        self.name = name
        self.__unit_price = __unit_price
        self.__qty = __qty

    def create_item(name):
        p = int(input("가격을 입력해주세요: "))
        q = int(input("수량을 입력해주세요: "))
        return Orderitem(name, p, q)

    def set_price(self, p):
        if p < 0:
            print("가격/수량은 0 이상이여야 합니다.")
        elif p >= 0:
            __unit_price = p
            return __unit_price
        
    def set_qty(self, q):
        if q < 0:
            print("가격/수량은 0 이상이여야 합니다.")
        elif q >= 0:
            __qty = q
            return __qty
        
    def total_price(self):
        t = self.__unit_price * self.__qty
        return t
    
    def __str__(self):
        return f"{self.name} * {self.__qty} = {self.total_price()}원"
    
    item1 = create_item("아메리카노")
    item2 = create_item("치즈케이크")

    print(item1)
    print(item2)
    total_sum = item1.total_price() + item2.total_price()
    print(f"총합: {total_sum}원")
```

```python
# 3번 문제
class MenuItem:
    def __init__(self, name, unit_price):
        self.name = name
        self.unit_price = unit_price

    def get_info(self):
        return f"{self.name} - {self.unit_price}원"

class Coffee(MenuItem):
    def __init__(self, name, unit_price, shot=1):
        super().__init__(name, unit_price)
        self.shot = shot

    def get_info(self):
        return f"{self.name}(샷 {self.shot}) - {self.unit_price}원"

class Dessert(MenuItem):
    def __init__(self, name, unit_price, size="M"):
        super().__init__(name, unit_price)
        self.size = size

    def get_info(self):
        return f"{self.name}({self.size}) - {self.unit_price}원"

items = [
    Coffee("아메리카노", 3000, 2),
    Dessert("치즈케이크", 5500, "L"),
    Coffee("에스프레소", 2500)  # 기본 샷 1 적용
]

print("--- 카페 메뉴 정보 ---")
for item in items:

    print(item.get_info())
```
